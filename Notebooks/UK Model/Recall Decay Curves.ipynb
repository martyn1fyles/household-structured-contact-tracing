```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
days_since_contact_occurred = np.array(range(12))
```

```python
recall_prob_95 = (0.95)**days_since_contact_occurred
recall_prob_90 = (0.9)**days_since_contact_occurred
```

```python
plt.plot(days_since_contact_occurred, recall_prob_90)
plt.plot(days_since_contact_occurred, recall_prob_95)
plt.ylim((0,1.1))
plt.title("Recall Probabilities")
plt.xlabel("Days Since Contact Occurred")
plt.ylabel("Recall Probabilities")
plt.legend(("10% recall decay per day", "5% recall decay per day"))
plt.savefig("Recall Model.pdf")
```
